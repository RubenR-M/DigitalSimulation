## Cadena de Markov, ejercicio

Supongamos que hay tres centros principales de camiones de una empresa.

Cada mes, la mitad de los que están en Boston y en Los Angeles, van a Chicago, 

la otra mitad permanece donde está y los camiones de Chicago se dividen 

igualmente entre Boston y L.A.



Si inicialmente la compañía tenía 100,200 y 300 camiones en Boston, Chicago y L.A. respectivamente, 

encontrar la distribuciones de camiones después de un mes y después de dos meses en las tres ciudades. 

```python
import numpy as np

# Matriz de Probabilidades de Transición
P = np.array([[1/2,1/2,0],[1/2,0,1/2],[0,1/2,1/2]])

# Estadp inicial
X_0 = [[100],[200],[300],"Inicio"]

Estados = []

Estados.append(X_0)

for i in range(1,13):
    estadoPrevio = np.array(Estados[i-1][0:3])
    nuevoEstado = np.matmul(P,estadoPrevio).round().tolist()
    nuevoEstado.append(f"mes {i}")
    Estados.append(nuevoEstado)
```
